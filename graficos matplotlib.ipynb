```python
from matplotlib import pyplot as plt
%matplotlib inline
import random
import numpy as np
import pandas as pd
import seaborn as sns
x = range(5)
x = np.array(x) # we have to convert to a numpy array, otherwise the plot
#does not perform the operation
fig, ax = plt.subplots(1, 2, figsize = (12, 5))# creates the figure with 
#subplots, and using size to adjust the sizes

#plt.plot(dados1, dados2) #pyplot erences the figure and the axis
print("tipo de ax =", type(ax))
print("conteudo de ax[0] =", ax[0])
print("conteudo de ax[1] =", ax[1])
ax[0].plot(x, x, label='eq_1', color="red")# creates the graph on the 0 axis
ax[0].plot(x, x**2, label='eq_2', color="yellow")#creates the graph on the 0 axis
ax[0].plot(x, x**3, label='eq_3', color="green")# creates the graph on the 0 axis
ax[0].set_xlabel('Eixo x')
ax[0].set_ylabel('Eixo y')
ax[0].set_title("Grafico 1")
ax[0].legend()
ax[1].plot(x, x, 'r--', label='eq_1') # creates the graph on the 0 axis
ax[1].plot(x**2, x, 'y--', label='eq_2') #creates the graph on the 0 axis
ax[1].plot(x**3, x, 'g--', label='eq_3') #creates the graph on the 0 axis
ax[1].set_xlabel('Novo Eixo x')
ax[1].set_ylabel('Novo Eixo y')
ax[1].set_title("Grafico 2")
ax[1].legend()
```

```python

x = range(5)
x = np.array(x) #temos que converter para um array numpy sesao o plot nao funcion
fig = plt.subplots(figsize=(12,5)) #criar figura sem eixo
plt.subplot(121) #adiciona um grid de subplot a figura: linha 1 e linha 2


plt.plot(x, x, label='eq_1')
plt.plot(x, x**2, label='eq_2')
plt.plot(x, x**3, label='eq_3')
plt.title("Grafico 1")
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()


plt.subplot(122) # adiciona um grid d subplot a figura: linha 1 e linha 2


plt.plot(x, x, label='rosa', color='pink')
plt.plot(x**2, x, label='cinza escuro', color='darkgrey')
plt.plot(x**3, x, label='laranja', color='orange')
plt.legend()
plt.title("Grafico das cores")
plt.xlabel('novo eixo x')
plt.ylabel('novo eixo y')
```

```python
import seaborn as sns

df = pd.read_csv('./arquivoscsv/cities.csv')

sns.set(style="whitegrid")
df_tips = sns.load_dataset('tips')

#biblioteca seaborn
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum)
sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len)

plt.figure(figsize=(10, 5))
ax=sns.barplot(x="day", y="total_bill", data=df_tips)
ax.axes.set_title("Vendas medias diarias", fontsize=14)
ax.set_xlabel("Dia", fontsize=14)
ax.set_ylabel("Venda media", fontsize=14)
ax.tick_params(labelsize=14)

#funo linear scatterplot() cria um grafico com alvos especificos 
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_tips, x="total_bill", y="tip")




df_tips.head()
print(df_tips.info())
df.head()
```
